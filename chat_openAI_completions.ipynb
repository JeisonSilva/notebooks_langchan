## Projeto para demonstrar a utilização do Chat Completions

#### Instação das bibliotecas

```polyglot-notebook
#r "nuget: OpenAI, 2.8.0"
#r "nuget: DotNetEnv, 3.1.1"

```

#### Carrega as configurações

```polyglot-notebook
using DotNetEnv;

var PathDocuments = System.Environment.GetFolderPath(System.Environment.SpecialFolder.MyDocuments);
var FilePath = System.IO.Path.Combine(PathDocuments, "estudos/langchan/.env");

Env.Load(FilePath);
string apiKey = Environment.GetEnvironmentVariable("API_KEY") ?? throw new InvalidOperationException("API_KEY not found in environment variables.");
```

#### Solicitação de uma pergunta usando o modelo ChatCompletion

```polyglot-notebook
using OpenAI;
using OpenAI.Chat;

var openAiClient = new OpenAIClient(apiKey);

var chatClient = openAiClient.GetChatClient(model: "gpt-3.5-turbo");
var result = await chatClient.CompleteChatAsync([
    new SystemChatMessage("Você é um professor de Astronomia para crianças de 11 anos com objetivo de tirar duvidas dos alunos"),
    new UserChatMessage("Como o universo foi criado?")
]);


```

#### Resultado do processamento da IA

```polyglot-notebook
Console.WriteLine(result.Value.Content[0].Text);
```
